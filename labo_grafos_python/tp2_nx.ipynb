```python
ws = pd.read_table('windsurfers.dat',skiprows=7,sep='\s+',header=None)
ws.shape
```

```python
ws
```

```python
ws_obs = ws.values[0:43,].astype(int)
for i in range(0,43): ws_obs[i,i]=0

ws_per = ws.values[43:86,]
for i in range(0,43): ws_per[i,i]=0
```

```python
ws_obs_bin = (ws_obs > 0)
ws_obs_bin
```

```python
G = nx.from_numpy_matrix(ws_obs,parallel_edges=False)
```

```python
G
```

```python
plt.subplot(131)
nx.draw(G, with_labels=True, font_weight='bold')
plt.subplot(132)
layout=nx.fruchterman_reingold_layout(G)
nx.draw(G,layout, with_labels=True, font_weight='bold')
plt.subplot(133)
layout=nx.spring_layout(G)
nx.draw(G,layout, with_labels=True, font_weight='bold')
```

```python
G.degree()
```

```python
G.edges(data=True)
```

```python
pos = nx.spring_layout(G)  # positions for all nodes
node_sizes = [20*x for x in list(G.degree().values())]
edge_colors = [x[2]['weight'] for x in G.edges(data=True)]


nodes = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='blue')
#nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')
edges = nx.draw_networkx_edges(G, pos, node_size=node_sizes, 
                           edgelist=G.edges(),edge_color=edge_colors,
                           edge_cmap=plt.cm.Blues, width=10)

# set alpha value for each edge
M = G.number_of_edges()
for i in range(M):
    edges[i].set_alpha(edge_colors[i])
    
#pc = mpl.collections.PatchCollection(edges, cmap=plt.cm.Blues)
#pc.set_array(edge_colors)
#plt.colorbar(pc)

plt.axis('off')
plt.show()
```
