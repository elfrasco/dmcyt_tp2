```python
ws = pd.read_csv('windsurfers.dat',skiprows=7,sep='\s+',header=None)
ws.shape
```

```python
ws_obs_bin_red = ig.Graph.Adjacency(ws_obs.tolist())
```

```python
ws_obs_bin_red.density(), ws_obs_bin_red.diameter()
```
